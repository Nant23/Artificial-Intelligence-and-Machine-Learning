```python
# Task 1

def convert_units(conversion_type, value):
    """
    Converts a value between different units of measurement.

    Parameters:
    conversion_type (int): The type of conversion selected by the user.
        1 - Meters to Feet
        2 - Feet to Meters
        3 - Kilograms to Pounds
        4 - Pounds to Kilograms
        5 - Liters to Gallons
        6 - Gallons to Liters
    value (float): The numeric value to be converted.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:
        return value * 3.28084  # meters to feet
    elif conversion_type == 2:
        return value / 3.28084  # feet to meters
    elif conversion_type == 3:
        return value * 2.20462  # kg to pounds
    elif conversion_type == 4:
        return value / 2.20462  # pounds to kg
    elif conversion_type == 5:
        return value * 0.264172  # liters to gallons
    elif conversion_type == 6:
        return value / 0.264172  # gallons to liters
    else:
        return None


def main():
    print("==== Unit Converter ====")
    print("Select Conversion Type:")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    try:
        choice = int(input("Enter your choice (1-6): "))

        if choice < 1 or choice > 6:
            print("Error: Invalid conversion type selected.")
            return

        value = float(input("Enter the value to convert: "))

        result = convert_units(choice, value)

        if result is not None:
            print(f"Converted value: {result:.4f}")
        else:
            print("Error: Conversion could not be performed.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")


if __name__ == "__main__":
    main()
```

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The maximum number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The minimum number in the list.
    """
    return min(numbers)


def main():
    print("==== Mathematical Operations on a List ====")
    print("Select Operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Enter your choice (1-4): "))

        if choice < 1 or choice > 4:
            print("Error: Invalid operation selected.")
            return

        user_input = input("Enter numbers separated by spaces: ")

        # Convert input string to list of floats
        numbers = list(map(float, user_input.split()))

        if len(numbers) == 0:
            print("Error: The list cannot be empty.")
            return

        if choice == 1:
            result = calculate_sum(numbers)
            print(f"Sum: {result}")
        elif choice == 2:
            result = calculate_average(numbers)
            print(f"Average: {result}")
        elif choice == 3:
            result = find_maximum(numbers)
            print(f"Maximum: {result}")
        elif choice == 4:
            result = find_minimum(numbers)
            print(f"Minimum: {result}")

    except ValueError:
        print("Error: Please enter valid numeric values.")


if __name__ == "__main__":
    main()
```

**Exercise on Nested List:**

```python
# 1. Flatten a Nested List:

def flatten(lst):
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list


# Example usage
numbers = [[1, 2], [3, 4], [5]]
result = flatten(numbers)
print(result)
```

```python
# 2. Accessing Nested List Elements:

def access_nested_element(lst, indices):
    element = lst
    for index in indices:
        element = element[index]
    return element

# Example usage:
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

result = access_nested_element(lst, indices)
print(result)
```

```python
# 3. Sum of All Elements in a Nested List:

def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # Recursively sum nested lists
        else:
            total += item
    return total

# Example usage:
nested_list = [[1, 2], [3, [4, 5]], 6]
print(sum_nested(nested_list))  # Output: 21
```

```python
# 4. Remove Specific Element from a Nested List:

def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

# Example usage:
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2
print(remove_element(lst, elem))
```

```python
# 5. Find the Maximum Element in a Nested List:

def find_max(lst):
    max_val = float('-inf')  # Start with the smallest possible value
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))  # Recursively find max in sublist
        else:
            max_val = max(max_val, item)  # Compare with current item
    return max_val

# Example usage:
nested_list = [[1, 2], [3, [4, 5]], 6]
print(find_max(nested_list))
```

```python
# 6. Count Occurrences of an Element in a Nested List:
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  # Recursively count in sublist
        elif item == elem:
            count += 1
    return count

# Example usage:
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2
print(count_occurrences(lst, elem))
```

```python
# 7. Flatten a List of Lists of Lists:

def deep_flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # Recursively flatten sublists
        else:
            flat_list.append(item)
    return flat_list

# Example usage:
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(nested_list))
```

```python
# 8. Nested List Average:

def average_nested(lst):
    def helper(lst):
        total = 0
        count = 0
        for item in lst:
            if isinstance(item, list):
                sub_total, sub_count = helper(item)
                total += sub_total
                count += sub_count
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = helper(lst)
    return total_sum / total_count if total_count != 0 else 0

# Example usage:
nested_list = [[1, 2], [3, 4], [5, 6]]
print(average_nested(nested_list))
```

```python
# Problem - 4: Matrix Operations:

# Given Matrices
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

# 1. Prove A.A−1 = I

A_inv = np.linalg.inv(A)     # Inverse of A
I = np.matmul(A, A_inv)         # A * A⁻¹

print("A inverse:\n", A_inv)
print("\nA * A⁻¹:\n", I)
```

```python
# Prove AB ̸= BA

AB = np.matmul(A, B)
BA = np.matmul(B, A)

print("AB:\n", AB)
print("\nBA:\n", BA)
```

```python
# Prove (AB)T = BTAT

left_side = (np.matmul(A, B)).T
right_side = np.matmul(B.T, A.T)

print("(AB)^T:\n", left_side)
print("\nB^T A^T:\n", right_side)
```

```python
# Solve Linear Equations Using Matrix Form

# Represent in Matrix Form AX=B
A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])

B_sys = np.array([-1, -3, 9])

A_sys_inv = np.linalg.inv(A_sys)
X = np.matmul(A_sys_inv, B_sys)

print("Solution using inverse method:\n", X)
print(f"x = {X[0]}\ny = {X[1]}\nz = {X[2]}")
```

```python
# Solve Directly Using np.linalg.inv()

X2 = np.linalg.solve(A_sys, B_sys)
print("Solution using np.linalg.solve():\n", X2)
print(f"x = {X2[0]}\ny = {X2[1]:.1f}\nz = {X2[2]}")
```

**Experiment: How Fast is Numpy?**

Element-wise Addition

```python
import time
import random

# Element-wise Addition using python lists:

# Create two lists of size 1,000,000
list1 = [random.random() for _ in range(1_000_000)]
list2 = [random.random() for _ in range(1_000_000)]

# Measure time for element-wise addition
start_time = time.time()

result_list = [a + b for a, b in zip(list1, list2)]

end_time = time.time()

print("Time taken using Python Lists:", end_time - start_time, "seconds")
```

```python
# Element-wise Addition using numpy arrays:

# Create two NumPy arrays of size 1,000,000
array1 = np.random.random(1_000_000)
array2 = np.random.random(1_000_000)

# Measure time for element-wise addition
start_time = time.time()

result_array = array1 + array2

end_time = time.time()

print("Time taken using NumPy Arrays:", end_time - start_time, "seconds")
```

Matrix Multiplication

```python
# Using Python lists
# Create 1000x1000 matrices
size = 1000
A = [[random.random() for _ in range(size)] for _ in range(size)]
B = [[random.random() for _ in range(size)] for _ in range(size)]

start = time.time()

# Matrix multiplication using nested loops
result = [[0]*size for _ in range(size)]

for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Matrix Multiplication (List):", end - start, "seconds")
```

```python
# Using numpy arrays

A_np = np.random.random((1000, 1000))
B_np = np.random.random((1000, 1000))

start = time.time()

result_np = np.dot(A_np, B_np)

end = time.time()

print("Matrix Multiplication (NumPy):", end - start, "seconds")
```
